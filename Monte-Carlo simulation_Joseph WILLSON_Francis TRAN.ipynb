# Monte-Carlo simulation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parameter Values
S0 = 50. # initial index level
K = 50. # strike price
T = 10.0 # time-to-maturity
r = 0.02 # riskless short rate
sigma = 0.2 # volatility
I = [1, 10, 100, 500, 1000, 2000, 3000, 5000, 10000, 100000, 1000000,100000000] # number of simulations

l = []

for i in range(len(I)):
    # Valuation Algorithm
    z = np.random.standard_normal(I[i]) # pseudorandom numbers
    ST = S0 * np.exp((r - 0.5 * sigma ** 2) * T + sigma * np.sqrt(T) * z)



    # index values at maturity
    hT = np.maximum(ST - K, 0) # inner values at maturity
    C0 = np.exp(-r * T) * np.sum(hT) / I[i] # Monte Carlo estimator

    l.append(C0)

    # Result Output
    print("Value of the Call Option using the Monte-Carlo simulation for N =", I[i], "is", round(C0,2))
    
    
```

### Above we implement the Monte-Carlo process to compute the price of the call option with different scenarios

```python
plt.plot(l)

plt.ylabel("Call option price")


plt.show()
```

### Above we have the trend wich move toward 16,35
